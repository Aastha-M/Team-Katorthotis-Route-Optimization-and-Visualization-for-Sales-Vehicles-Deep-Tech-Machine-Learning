```
class env_map:
    def __init__(self):
        self.jun_link=[[2,4],[1,3],[2,4,5],[1,3,6],[3,6],[4,5]]
        self.jun_dist=[[10,15],[10,5],[5,7,3],[15,7,11],[3,7],[7,11]]

```

```
class path:
    def __init__(self):
        self.pth=[]
        self.pth_dist=[]
        self.dist=0
    def load_pth(self,p):
        self.pth=p.pth
        self.pth_dist=p.pth_dist
        self.dist=p.dist
        self.pos=self.pth[len(self.pth)-1]
    def mov(self,j,d):
        self.pth.append(j)
        self.dist+=d
        self.pth_dist.append(self.dist)
        self.pos=j
```

```
def term_test(paths,des):
    #destination check state
    sols=[]
    i=0
    #for i in range(len(paths)):
    while(i<len(paths)):
        p=paths[i]
        if(p.pos==des):
            sols.append(p)
            paths.remove(p)
            i=i-1
            continue
        curr_pos=p.pos
        curr_dis=p.dist
        j=0
        #for j in range(len(paths)):
        while(j<len(paths)):
            if(i==j):
                continue
            p2=paths[j]
            pth=p2.pth
            if(curr_pos in pth):
                ind=pth.index[curr_pos]
                dist2=p2.pth_dist[ind]
                if(dist2>curr_dist):
                    paths.remove(p)
                    j=j-1
                    i=i-1
                    break
                else:
                    paths.remove(p2)
                    j=j-1
                    if(ind<i):
                        i=i-1
            j=j+1
            print('j'+str(j))
            print('len(paths):'+str(len(paths)))
        i=i+1
        print('i'+str(i))
        print('len(paths):'+str(len(paths)))    
    return paths,sols
```

```
def combine_sol(sols,sol):
    d1=sol[0].dist
    s=sol[0]
    for p in sols:
        d2=p.dist
        if(d1>d2):
            s=p
    sol=[s]
    return sol
```
